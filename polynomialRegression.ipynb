```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("polynomial_data.csv")

plt.figure(figsize=(10, 6))
plt.scatter(df["x_train"], df["y_train"], color="teal", label="Loaded Data", alpha=0.8)
plt.title("Loaded Polynomial Regression Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

print("Loaded data shape:")
```

```python
x = df["x_train"].values.reshape(-1, 1)
y = df["y_train"].values.reshape(-1, 1)

meanX = np.mean(x)
stdX = np.std(x)
x = (x - meanX)/stdX

x2 = x ** 2
x3 = x** 3
x4 = x**4
x5 = x**5
x6 = x**6



x_train = np.array([x.flatten(), x2.flatten(), x3.flatten(), x4.flatten(), x5.flatten(),x6.flatten(),]).T

m = x_train.shape[0]


w = np.array([[2.0], [1.0], [0.9], [3.0], [5.0], [1.0]])
b = 0
alpha = 0.02
allCost = []

for i in range(50000):
    y_pred = x_train @ w + b  
    cost = (1/(2*m)) * np.sum((y - y_pred) ** 2)
    

    allCost.append(cost)
    
    
    dw = (1 / m) * (x_train.T @ (y_pred - y))
    db = (1 / m) * np.sum(y_pred - y)
    
    w = w - alpha * dw
    b = b -  alpha * db
    

print("Weights:\n", w)
print("Bias:", b)
print(allCost[-1])
plt.scatter(x, df["y_train"], color="teal", label="Loaded Data", alpha=0.8)
plt.plot(x, y_pred)
plt.show()
plt.plot(allCost)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost over iterations")
plt.grid(True)
plt.show()
```
